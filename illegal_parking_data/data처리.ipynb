```python
import requests
import pprint
import pandas as pd



PATH="2019불법/"
nam=pd.read_csv(PATH+"남구2019.csv",encoding='utf-8-sig')
nam=nam[['단속일자','단속시간','단속장소']]
nam['lat']=0
nam['lng']=0


```

```python
for i in range(0,len(nam)):
    location=("대구광역시 남구"+str(nam.loc[i][2]))
    
    URL = 'http://maps.googleapis.com/maps/api/geocode/json?sensor=false&language=ko&address={}'.format(location)

    URL = 'https://maps.googleapis.com/maps/api/geocode/json?key=AIzaSyAUnVkEyQBLy5NrJAZC_km_WMh3Eypl7RQ&sensor=false&language=ko&address={}'.format(location)
    
    #JSON 파싱
    response = requests.get(URL)
    
    data = response.json()
    
    #lat, lon 추출
    lat = data['results'][0]['geometry']['location']['lat']
    lng = data['results'][0]['geometry']['location']['lng']

    nam.loc[i,'lat']=lat
    nam.loc[i,'lng']=lng
    print(nam.loc[i])
```

```python
import requests
import pprint
import pandas as pd



PATH="2019불법/"
data=pd.read_csv(PATH+"수성구2019.csv",encoding='utf-8-sig')
data=data[['단속일자','단속시간','단속장소']]
data['lat']=0
data['lng']=0

data
```

```python
for i in range(16782,len(data)):
    location=("대구광역시 수성구"+str(data.loc[i][2]))
    
    URL = 'http://maps.googleapis.com/maps/api/geocode/json?sensor=false&language=ko&address={}'.format(location)

    URL = 'https://maps.googleapis.com/maps/api/geocode/json?key=AIzaSyAY4EwEirPVH_vPnflzdPXkiSwSoEUJgWU&sensor=false&language=ko&address={}'.format(location)
    
    #JSON 파싱
    response = requests.get(URL)
    
    data1 = response.json()
    #lat, lon 추출
    lat = data1['results'][0]['geometry']['location']['lat']
    lng = data1['results'][0]['geometry']['location']['lng']

    data.loc[i,'lat']=lat
    data.loc[i,'lng']=lng
    print(data.loc[i])
```
