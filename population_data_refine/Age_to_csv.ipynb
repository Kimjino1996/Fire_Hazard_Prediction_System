```python
import pandas as pd
from pandas import DataFrame
xlsx = pd.read_excel('C:\\Users\kimjino\Desktop\data2.xlsx')
```

```python
name='삼덕동'
df = DataFrame(columns=("나이",name))
df2=[]
for i in range(0,21):
    k=6*i
  # print("i 값은",k)
    df.loc[i]=xlsx['Unnamed: 0'][2+k],xlsx[name][2+k]

for z in xlsx.columns:
    #print (k)
    if 'U' in z or '합'in z:
        print(' ')
    else :
        print('hi')
        for i in range(0,21):
            k=6*i
          # print("i 값은",k)
            print(xlsx[z][2+k])
            df2.append(xlsx[z][2+k])
        
        df[z]=df2
        df2=[]

print(df.columns)
df
```
