```python
import pandas as pd
```

```python
data=pd.read_csv("apt.csv",encoding='utf-8-sig')
data['Escore']=0
data
```

```python
el=pd.read_csv("electricity_score.csv",encoding="utf-8-sig")
el
```

```python
for i in range(len(data)):
    for j in range(len(el)):
        if(data.loc[i,'location']==el.loc[j,'대지위치']):
            data.loc[i,'Escore']=el.loc[j,'score']
```

```python
data
```

```python
data.to_csv('apt_2.csv',encoding="utf-8-sig")
```

```python
for i in range(len(data)):
    if(data.loc[i,'Gscore']==0):
        data.loc[i,'Gscore']=5
    if(data.loc[i,'Escore']==0):
        data.loc[i,'Escore']=5
```

```python
data.to_csv('apt_3.csv',encoding="utf-8-sig")
```

```python
data['노후정도점수']=0
apt=data
```

```python
cnt = 0

for i in apt['연식']:
    if i>30:
        apt.loc[cnt,'노후정도점수'] += 20
    elif 20<=i and i<30:
        apt.loc[cnt,'노후정도점수'] += 15
    elif 10<=i and i<20:
        apt.loc[cnt,'노후정도점수'] += 10
    elif 0<=i and i<10:
        apt.loc[cnt,'노후정도점수'] += 5
        
    cnt += 1
```

```python
apt
```

```python
apt.to_csv("apt_new.csv",encoding="utf-8-sig")
```
