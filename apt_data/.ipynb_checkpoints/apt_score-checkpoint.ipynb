#  APT 데이터

```python
import pandas as pd
```

```python
apt = pd.read_csv("apt.csv")
apt
```

```python
apt.columns
```

```python
apt = apt.drop(['Unnamed: 0', '사  업 주체별','연도','관리별','관리실 전  화'], axis=1)
```

# 연식 구하기

```python
apt['연식']=0
```

```python
cnt = 0
for i in apt['건축 연도']:
    apt.loc[cnt,'연식'] = 2020-i
    cnt += 1
```

```python
apt
```

# Score 추가

```python
apt['score'] = 0 
```

```python
cnt = 0
for i in apt['연식']:
    if i>30:
        apt.loc[cnt,'score'] += 20
    elif 20<=i and i<30:
        apt.loc[cnt,'score'] += 15
    elif 10<=i and i<20:
        apt.loc[cnt,'score'] += 10
    elif 0<=i and i<10:
        apt.loc[cnt,'score'] += 5
        
    cnt += 1
```

```python
apt
```

```python
apt.to_csv("new_apt.csv", encoding='utf-8-sig')
```

# Folium에 표시

```python
import folium
```

```python
latitude_ = apt['lat'].mean()
longitude_ = apt['lng'].mean()

m = folium.Map([latitude_,longitude_],zoom_start=12)
```

```python
folium.CircleMarker(
      [35.860484, 128.581216],
      radius=20,
      color='red',
      fill=True,
      popup = '한마음'
    ).add_to(m)
```

```python
for i in range(0,len(apt)):
    lat = float(apt.loc[i, 'lat'])
    lng = float(apt.loc[i, 'lng'])
    score = int(apt.loc[i, 'score'])
    
    folium.CircleMarker(
      [lat, lng],
      radius=score/5,
      color='red',
      fill=True,
      popup = apt.loc[i, '아파트명']
    ).add_to(m)
    
    print(lat,lng,score)
```

```python
m.save("map.html")
```
