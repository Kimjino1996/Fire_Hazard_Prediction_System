# Data(노후아파트연도) 정리

```python
import pandas as pd
```

```python
apt = pd.read_excel('./노후아파트연도x.xlsx')
```

```python
apt.head(5)
```

```python
apt = apt.drop('Unnamed: 2',axis=1)
apt = apt.drop('위      치.1',axis=1)
```

```python
apt.head(5)
```

```python
apt.columns
```

```python
apt['latitude'] = 0
apt['longitude'] = 0
```

```python
apt_loc = apt['위      치']
```

```python
cnt = 0 
latitude = []
longitude = []

for location in apt_loc:
    location = '대구' + location
    
    URL = 'http://maps.googleapis.com/maps/api/geocode/json?sensor=false&language=ko&address={}'.format(location)
    URL = 'https://maps.googleapis.com/maps/api/geocode/json?key=AIzaSyAUnVkEyQBLy5NrJAZC_km_WMh3Eypl7RQ&sensor=false&language=ko&address={}'.format(location)
    
    response = requests.get(URL)
    
    data = response.json()
    
    #print(data)
    
    if(data['status']=='ZERO_RESULTS'):
        location = '대구' + apt.iat[cnt, 1] + '아파트'
        
        URL = 'http://maps.googleapis.com/maps/api/geocode/json?sensor=false&language=ko&address={}'.format(location)
        URL = 'https://maps.googleapis.com/maps/api/geocode/json?key=AIzaSyAUnVkEyQBLy5NrJAZC_km_WMh3Eypl7RQ&sensor=false&language=ko&address={}'.format(location)
    
        response = requests.get(URL)
    
        data = response.json()
        
    
    lat = data['results'][0]['geometry']['location']['lat']
    lng = data['results'][0]['geometry']['location']['lng']
    
    print(location,"-",lat,'-',lng)
    
    apt.iat[cnt,10] = float(lat)
    apt.iat[cnt,11] = float(lng)
    
    latitude.append(lat)
    longitude.append(lng)
    
    
    cnt += 1
    
print("complete")
```

```python
apt = apt.drop(['latitude', 'longitude'], axis=1)
```

```python
apt['lat'] = latitude
apt['lng'] = longitude
```

```python
apt
```

```python
apt.to_csv("apt", encoding='utf-8-sig')
```

# Folium으로 표시

```python
import folium
```

```python
latitude_ = apt['lat'].mean()
longitude_ = apt['lng'].mean()

m = folium.Map([latitude_,longitude_],zoom_start=12)
m
```

```python
for i in apt.index:
    sub_lat = apt.loc[i,'lat']
    sub_lng = apt.loc[i,'lng']
    
    title = apt.loc[i, '아파트명']
    
    folium.Marker([sub_lat,sub_lng],tooltip = title).add_to(m)

m.save("apt_map.html")
```
