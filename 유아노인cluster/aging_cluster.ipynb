```python
import pandas as pd
import folium
```

```python
data=pd.read_csv("aging_population.csv",encoding="utf-8-sig")
data
```

```python
data['유아노인']=0
```

```python
for i in range(len(data)):
    data.loc[i,'유아노인']=data.loc[i,'0∼4세']+data.loc[i,'5∼9세']
    data.loc[i,'유아노인']+=data.loc[i,'65∼69세']+int(data.loc[i,'70∼74세'])
    data.loc[i,'유아노인']+=data.loc[i,'75∼79세']+data.loc[i,'80∼84세']
    data.loc[i,'유아노인']+=data.loc[i,'85∼89세']+data.loc[i,'90∼94세']
    data.loc[i,'유아노인']+=data.loc[i,'95∼99세']+data.loc[i,'100세이상']
    
```

```python
data
```

```python
map_osm=folium.Map(location=[35.798838,128.583052],zoom_start=14)

for i in range(len(data)):
    lat1=data.loc[i,'lat']
    lng1=data.loc[i,'lng']
    if data.loc[i,'유아노인']>10000:
        folium.CircleMarker([lat1,lng1],radius=10,popup=data.loc[i,'유아노인'],color='#1864ab', fill_color='#1864ab').add_to(map_osm)
    elif data.loc[i,'유아노인']>7000:
        folium.CircleMarker([lat1,lng1],radius=10,popup=data.loc[i,'유아노인'],color='#228be6', fill_color='#228be6').add_to(map_osm)
    elif data.loc[i,'유아노인']>4000:
        folium.CircleMarker([lat1,lng1],radius=10,popup=data.loc[i,'유아노인'],color='#4dabf7', fill_color='#4dabf7').add_to(map_osm)
    elif data.loc[i,'유아노인']>1000:
        folium.CircleMarker([lat1,lng1],radius=10,popup=data.loc[i,'유아노인'],color='#a5d8ff', fill_color='#a5d8ff').add_to(map_osm)
    elif data.loc[i,'유아노인']<1000:
        folium.CircleMarker([lat1,lng1],radius=10,popup=data.loc[i,'유아노인'],color='#e7f5ff', fill_color='#e7f5ff').add_to(map_osm)
```

```python
map_osm.save('유아노인cluster.html')
```
